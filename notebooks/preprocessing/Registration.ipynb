```
import nibabel as nib
import numpy as np
from PIL import Image
from os import listdir
from os.path import isfile, join, basename, split
from tqdm import tqdm
```

Creating the directories

```
!mkdir drive/My\ Drive/DLMI\ Project/Data/CT_slices_nii
!mkdir drive/My\ Drive/DLMI\ Project/Data/T2_slices_nii
```

Get the paths for the images of CT and T2 images

```
dir_path_CT = 'drive/My Drive/DLMI Project/Data/CT_slices/'
dir_path_T2 = 'drive/My Drive/DLMI Project/Data/T2_slices/'
image_paths_CT = sorted([join(dir_path_CT, file) for file in listdir(dir_path_CT) if isfile(join(dir_path_CT, file))])
image_paths_T2 = sorted([join(dir_path_T2, file) for file in listdir(dir_path_T2) if isfile(join(dir_path_T2, file))])
```

Convert the CT and T2 images

```
for path in tqdm(image_paths_CT):
    img = np.asarray(Image.open(path).convert('L'))
    img_nii = nib.Nifti1Image(img, affine=np.eye(4))
    new_path = join(split(path)[0]+'_nii', split(path)[1])[:-3]+'nii'
    nib.save(img_nii, new_path)
```

```
for path in tqdm(image_paths_T2):
    img = np.asarray(Image.open(path).convert('L'))
    img_nii = nib.Nifti1Image(img, affine=np.eye(4))
    new_path = join(split(path)[0]+'_nii', split(path)[1])[:-3]+'nii'
    nib.save(img_nii, new_path)
```
